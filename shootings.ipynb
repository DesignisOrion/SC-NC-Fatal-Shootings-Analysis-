# Fatal Shootings within the Carolinas

After the verdict of Breonna Taylor many addressed their concerns via the media. It made me want to take a more in depth look of the fatal shootings within the states of North and South Carolina. 




```python
import numpy as np
import pandas as pd


```

# Overview of fatal shootings within the United States

Here is all the data of the fatal shootings within United States. 2015 - 2020.

```python
shootings = pd.read_csv('fatal-police-shootings-data.csv')
shootings
```

# South Carolina
Here is the insight to fatal deaths within South Carolina. 

```python
stateSC = shootings.loc[shootings['state']=='SC']
stateSC.to_csv('stateSC.csv')
stateSC
```

# Males in South Carolina
Insight of males within South Carolina involved in fatal shootings

```python
malesSC= shootings.loc[(shootings['state'] == 'SC') & (shootings['gender'] == 'M')]
malesSC.to_csv('malesSC.csv')
malesSC
```

# Females in South Carolina
Insight of females within South Carolina involved in fatal shootings

```python
femaleSC = shootings.loc[(shootings['state'] == 'SC') & (shootings['gender'] == 'F')]
femaleSC.to_csv('femaleSC.csv')
femaleSC
```
